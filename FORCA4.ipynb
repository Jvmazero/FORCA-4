```python
import tkinter as tk
from tkinter import messagebox

class JogoDaForca:
    def __init__(self):
        self.palavras = ["python", "javascript", "ruby", "java", "html", "css"]
        self.palavra_secreta = ""
        self.letras_corretas = []
        self.letras_erradas = []
        self.tentativas_restantes = 6

        self.nova_palavra()
        self.criar_interface()

    def nova_palavra(self):
        self.palavra_secreta = random.choice(self.palavras)
        self.letras_corretas = []
        self.letras_erradas = []
        self.tentativas_restantes = 6

    def mostrar_palavra_oculta(self):
        palavra_mostrada = ""
        for letra in self.palavra_secreta:
            if letra in self.letras_corretas:
                palavra_mostrada += letra + " "
            else:
                palavra_mostrada += "_ "
        return palavra_mostrada.strip()

    def criar_interface(self):
        self.root = tk.Tk()
        self.root.title("Jogo da Forca")

        self.lbl_palavra = tk.Label(self.root, text=self.mostrar_palavra_oculta(), font=("Arial", 24))
        self.lbl_palavra.pack(pady=10)

        self.lbl_letras_usadas = tk.Label(self.root, text="Letras usadas: ", font=("Arial", 16))
        self.lbl_letras_usadas.pack()

        self.lbl_tentativas = tk.Label(self.root, text=f"Tentativas Restantes: {self.tentativas_restantes}", font=("Arial", 16))
        self.lbl_tentativas.pack()

        self.entry_palpite = tk.Entry(self.root, font=("Arial", 16))
        self.entry_palpite.pack(pady=10)

        self.btn_palpite = tk.Button(self.root, text="Fazer Palpite", command=self.verificar_palpite)
        self.btn_palpite.pack()

        self.btn_reiniciar = tk.Button(self.root, text="Reiniciar Jogo", command=self.reiniciar_jogo)
        self.btn_reiniciar.pack()

        self.root.mainloop()

    def verificar_palpite(self):
        palpite = self.entry_palpite.get().lower()
        self.entry_palpite.delete(0, tk.END)

        if len(palpite) != 1:
            messagebox.showwarning("Aviso", "Por favor, digite apenas uma letra como palpite.")
            return

        if palpite in self.letras_corretas or palpite in self.letras_erradas:
            messagebox.showinfo("Aviso", "Você já tentou essa letra.")
            return

        if palpite in self.palavra_secreta:
            self.letras_corretas.append(palpite)
            self.lbl_palavra.config(text=self.mostrar_palavra_oculta())
            if self.verificar_palavra_completa():
                messagebox.showinfo("Parabéns", f"Você ganhou! A palavra secreta era '{self.palavra_secreta}'.")
                self.reiniciar_jogo()
        else:
            self.letras_erradas.append(palpite)
            self.lbl_letras_usadas.config(text="Letras usadas: " + ", ".join(self.letras_erradas))
            self.tentativas_restantes -= 1
            self.lbl_tentativas.config(text=f"Tentativas Restantes: {self.tentativas_restantes}")
            if self.tentativas_restantes == 0:
                messagebox.showinfo("Game Over", f"Você perdeu! A palavra secreta era '{self.palavra_secreta}'.")
                self.reiniciar_jogo()

    def verificar_palavra_completa(self):
        for letra in self.palavra_secreta:
            if letra not in self.letras_corretas:
                return False
        return True

    def reiniciar_jogo(self):
        self.nova_palavra()
        self.lbl_palavra.config(text=self.mostrar_palavra_oculta())
        self.lbl_letras_usadas.config(text="Letras usadas: ")
        self.lbl_tentativas.config(text=f"Tentativas Restantes: {self.tentativas_restantes}")

# Exemplo de uso:
import random
jogo = JogoDaForca()
```
